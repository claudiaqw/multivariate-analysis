```r
library(tidyverse)
library(dplyr)
library(GGally) # ggplot2-based visualization of correlations
library(factoextra)
```

```r
original_sales = read.csv(file = "HomeSalesData.csv", header=T, sep=",")
glimpse(original_sales)
```

```r
sum(is.na(original_sales))
```

```r
sales <- select(original_sales, -c('id', 'date'))
head(sales)
```

```r
price <- original_sales[c('id', 'price')]
head(price)
```

```r
sales <- select(sales, -c('price'))
head(sales)
```

```r
summary(sales)
```

```r
boxplot(sales, las=2, col="darkblue")
```

```r
boxplot(sales$sqft_lot, las=2, col="darkblue")
```

```r
R = cor(sales)   # correlation matrix
R
```

```r
library(ggplot2)
plot(original_sales[,6], original_sales$price, type='p', col='blue', xlab = names(original_sales)[3], ylab = 'price' )
```

```r
head(original_sales)
```

```r
for(i in 4:21){
  plot(original_sales[,i], original_sales$price, type='p', col='blue', xlab = names(original_sales)[i], ylab = 'price' )
}
```

```r
#5. Aplicar herramienta PCA e interpretar: 2 puntos
pca = prcomp(sales, scale=T)
# pca = princomp(nba, cor=T) # the same, but using SVD instead of eigen decomposition 
summary(pca)
```

```r
eigen(R)
```

```r
screeplot(pca,main="Screeplot",col="blue",type="barplot",pch=18)
```

```r
fviz_contrib(pca, choice = "var", axes = 2)
```
